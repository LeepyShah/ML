<h3>15.Develop a Python program to implement DBSCAN using the sklearn library. Use a 2D dataset (ex.
make_moons) for clustering.</h3>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
# 1) Data
X, _ = make_moons(n_samples=300, noise=0.08, random_state=42)
X = StandardScaler().fit_transform(X)
```

```python
# 2) DBSCAN
db = DBSCAN(eps=0.25, min_samples=5)
labels = db.fit_predict(X)
```

```python
# 3) Plot
plt.figure(figsize=(6,5))
plt.scatter(X[:,0], X[:,1], c=labels)
plt.title("DBSCAN on make_moons")
plt.xlabel("x1"); plt.ylabel("x2")
plt.show()
```

```python
print("Clusters (excluding noise):", len(set(labels)) - (1 if -1 in labels else 0))
print("Noise points:", np.sum(labels == -1))
```

<h3>17.Use DBSCAN to cluster customers based on their purchasing behaviour (ex. frequency of purchases, average
transaction amount)(Generate random dataset with min. 100 customers). Develop a program that analyses
the clusters to identify different customer segments for targeted marketing</h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(7)
```

```python
# 1) Generate customers (3 patterns mixed)
n = 150
freq = np.concatenate([
    np.random.normal(30, 6, n//3),   # frequent buyers
    np.random.normal(15, 5, n//3),   # regular
    np.random.normal(6,  2, n - 2*(n//3))  # occasional
]).clip(1, None)
amt = np.concatenate([
    np.random.normal(2500, 500, n//3),   # high amount
    np.random.normal(1500, 400, n//3),   # medium
    np.random.normal(700,  200, n - 2*(n//3))  # low
]).clip(100, None)

df = pd.DataFrame({
    "purchase_frequency": freq.round(0),
    "avg_transaction_amount": amt.round(0)
})
```

```python
# 2) Scale + DBSCAN
X = StandardScaler().fit_transform(df)
db = DBSCAN(eps=0.5, min_samples=5)
labels = db.fit_predict(X)
df["cluster"] = labels
```

```python
# 3) Plot
plt.figure(figsize=(6,5))
plt.scatter(df["purchase_frequency"], df["avg_transaction_amount"], c=df["cluster"])
plt.title("DBSCAN Customer Segments")
plt.xlabel("Purchase Frequency")
plt.ylabel("Avg Transaction Amount")
plt.show()
```

```python
# 4) Segment summary
summary = df.groupby("cluster")[["purchase_frequency","avg_transaction_amount"]].agg(["count","mean","median"]).round(2)
print(summary)

print("\nMarketing idea:")
print("- High frequency + high amount cluster => VIP / loyalty offers")
print("- Medium cluster => bundles / cross-sell")
print("- Low cluster or noise => reactivation discounts")
```

<h3>18.Generate a data frame containing two features(Spending,features) for 100 data. Apply DBSCAN and do
parametric analysis for eps value and show the impact on the performance of model via graph.</h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(42)
```

```python
#DataFrame with 2 features for 100 points
df=pd.DataFrame({
    "spending":np.random.normal(5000,1800,100).clip(500,None),
    "visits":np.random.normal(12,5,100).clip(1,None)
})
```

```python

X = StandardScaler().fit_transform(df)

eps_values = np.linspace(0.1, 1.2, 15)
clusters_list = []
noise_list = []
```

```python
# 2) Vary eps
for eps in eps_values:
    labels = DBSCAN(eps=float(eps), min_samples=5).fit_predict(X)
    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise = np.sum(labels == -1)
    clusters_list.append(n_clusters)
    noise_list.append(n_noise)
```

```python

# 3) Graphs
plt.figure(figsize=(6,4))
plt.plot(eps_values, clusters_list, marker="o")
plt.title("Effect of eps on number of clusters")
plt.xlabel("eps")
plt.ylabel("# clusters")
plt.show()

plt.figure(figsize=(6,4))
plt.plot(eps_values, noise_list, marker="o")
plt.title("Effect of eps on noise points")
plt.xlabel("eps")
plt.ylabel("# noise points")
plt.show()
```

<h3>19.Generate a random dataset with 4 columns and use 1st and 3rd column as data to implement DBSCAN.
What if eps is changed from 1 to 5? How it will affect the performance of the model? What is the impact of
minPts on the performance of the model? Justify it with graph by changing minPts</h3>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

np.random.seed(1)
```

```python
# 1) Random data: 250 rows, 4 cols
data = np.random.randint(0, 50, size=(250, 4))
X = data[:, [0, 2]]  # 1st and 3rd column
X = StandardScaler().fit_transform(X)
```

```python
# ----- (A) eps from 1 to 5 -----
eps_values = [1,2,3,4,5]
plt.figure(figsize=(12,6))
for i, eps in enumerate(eps_values, start=1):
    labels = DBSCAN(eps=eps, min_samples=5).fit_predict(X)
    plt.subplot(2,3,i)
    plt.scatter(X[:,0], X[:,1], c=labels)
    plt.title(f"eps={eps}, minPts=5")
plt.tight_layout()
plt.show()
```

```python
# ----- (B) Impact of minPts (min_samples) -----
minPts_values = [3,5,10,15]
plt.figure(figsize=(10,8))
for i, m in enumerate(minPts_values, start=1):
    labels = DBSCAN(eps=2, min_samples=m).fit_predict(X)
    plt.subplot(2,2,i)
    plt.scatter(X[:,0], X[:,1], c=labels)
    plt.title(f"eps=2, minPts={m}")
plt.tight_layout()
plt.show()

print("Understanding:")
print("- Increasing eps => points connect more => fewer clusters, less noise")
print("- Increasing minPts => stricter density => more noise, fewer clusters")
```

<h3>Generate random dataset for mall_customer dataset with features like Customer, Genre, Age, Annual
Imcome, Spending Score. Consider minimum 200 random samples for analysis. Apply DBSCAN as well as
agglomerative clustering and generate the dendrogram. Plot points in each cluster for DBSCAN with
different colours</h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import DBSCAN, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, dendrogram

np.random.seed(42)
```

```python
# 1) Generate mall dataset (>=200)
n = 220
df = pd.DataFrame({
    "Customer": np.arange(1, n+1),
    "Genre": np.random.choice(["Male", "Female"], size=n),
    "Age": np.random.randint(18, 70, size=n),
    "Annual_Income": np.random.randint(15000, 150000, size=n),
    "Spending_Score": np.random.randint(1, 100, size=n)
})
```

```python
# Encode Genre (Male=0, Female=1)
df["Genre_code"] = (df["Genre"] == "Female").astype(int)
```

```python
# Use numeric features for clustering
features = ["Genre_code", "Age", "Annual_Income", "Spending_Score"]
X = StandardScaler().fit_transform(df[features])
```

```python
# 2) DBSCAN
db = DBSCAN(eps=0.7, min_samples=5)
df["dbscan_cluster"] = db.fit_predict(X)
```

```python
# DBSCAN plot (2D view: income vs spending)
plt.figure(figsize=(7,5))
plt.scatter(df["Annual_Income"], df["Spending_Score"], c=df["dbscan_cluster"])
plt.title("DBSCAN Clusters (Annual Income vs Spending Score)")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```

```python
# 3) Agglomerative clustering
agg = AgglomerativeClustering(n_clusters=4, linkage="ward")
df["agg_cluster"] = agg.fit_predict(X)
```

```python
# 4) Dendrogram (Agglomerative)
Z = linkage(X, method="ward")
plt.figure(figsize=(10,5))
dendrogram(Z, truncate_mode="lastp", p=25)  # show last 25 merges (clean graph)
plt.title("Dendrogram (Agglomerative - Ward)")
plt.xlabel("Cluster")
plt.ylabel("Distance")
plt.tight_layout()
plt.show()

print("DBSCAN clusters (excluding noise):", len(set(df["dbscan_cluster"])) - (1 if -1 in df["dbscan_cluster"].values else 0))
print("DBSCAN noise points:", (df["dbscan_cluster"] == -1).sum())
print("\nDBSCAN summary:")
print(df.groupby("dbscan_cluster")[["Age","Annual_Income","Spending_Score"]].mean().round(2))
```

<h3>Q18 – eps analysis (impact of eps)</h3>

```python
# STEP 1
df = pd.DataFrame({
    "spending": np.random.randint(1000,10000,100),
    "visits": np.random.randint(1,30,100)
})

X = StandardScaler().fit_transform(df)

eps_values = [0.2,0.4,0.6,0.8,1]

clusters=[]

for e in eps_values:
    labels = DBSCAN(eps=e, min_samples=5).fit_predict(X)
    clusters.append(len(set(labels))- (1 if -1 in labels else 0))

plt.plot(eps_values, clusters)
plt.title("Impact of eps")
plt.xlabel("eps")
plt.ylabel("No. of clusters")
plt.show()
```
