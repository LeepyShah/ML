<h3>58.Consider a dataset HousingData.csv with columns num_rooms, area, age, and a target column price.
Write a Python code to fit a Lasso regression model using num_rooms, area, and age as independent
variables to predict price. Use the complete dataset to train the model. Plot the coefficients of the
features and create a residual plot.
</h3>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Load data
df = pd.read_csv("Housing.csv")
```

```python
df.head()
```

```python
#it does NOT contain num_rooms or age 
#Rename bedrooms → num_rooms  &  stories → age
df.rename(columns={
    'bedrooms': 'num_rooms',
    'stories': 'age'  
}, inplace=True)
```

```python
# Select features and target
X = df[['num_rooms', 'area', 'age']]
y = df['price']
```

```python
# Build Lasso model
model = Pipeline([
    ('scaler', StandardScaler()),
    ('lasso', Lasso(alpha=0.1))
])
```

```python
# Train on complete dataset
model.fit(X, y)
```

```python
# Predictions
y_pred = model.predict(X)
residuals = y - y_pred
```

```python
# Get coefficients
coeff = model.named_steps['lasso'].coef_
features = X.columns
```

```python
# -------------------------
# Plot feature coefficients
# -------------------------
plt.figure(figsize=(6,4))
plt.bar(features, coeff)
plt.axhline(0)
plt.title("Lasso Feature Coefficients")
plt.ylabel("Coefficient Value")
plt.show()

# -------------------------
# Residual plot
# -------------------------
plt.figure(figsize=(6,4))
plt.scatter(y_pred, residuals)
plt.axhline(0)
plt.xlabel("Predicted Price")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```

<h3>60.Consider a dataset WineQuality.csv with various chemical properties of wine and a target column
quality. Write a Python code to fit a Lasso regression model using all available features to predict
quality. Use the complete dataset to train the model. Plot the coefficients of the features and create
a residual plot.</h3>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load data
df = pd.read_csv("WineQuality.csv")

target = "quality"
X = df.drop(columns=[target])
y = df[target]
```

```python
# -----------------------------
# 2) Preprocess + Lasso model
#    - Scale numeric features (important for Lasso)
#    - One-hot encode categorical (if any)
# -----------------------------
#Find numeric & categorical columns
num_cols = X.select_dtypes(include=["number"]).columns
cat_cols = X.select_dtypes(exclude=["number"]).columns

preprocess = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols)
    ],
    remainder="drop"
)
alpha = 0.05  # change if needed (e.g., 0.01, 0.1, 1.0)
model = Pipeline([
    ("prep", preprocess),
    ("lasso", Lasso(alpha=alpha, max_iter=10000))
])
```

```python
# Train on complete dataset
model.fit(X, y)
```

```python
# Predictions + residuals
y_pred = model.predict(X)
residuals = y - y_pred
```

```python
print("Alpha:", alpha)
print("RMSE:", np.sqrt(mean_squared_error(y, y_pred)))
print("R^2:", r2_score(y, y_pred))
```

```python
# -----------------------------
# 3) Coefficients + feature names
# -----------------------------
feature_names = model.named_steps["prep"].get_feature_names_out()
coefs = model.named_steps["lasso"].coef_
```

```python
# -----------------------------
# 4) Plot coefficients
# -----------------------------
plt.figure(figsize=(10, 4))
plt.bar(feature_names, coefs)
plt.axhline(0, linewidth=1)
plt.title("Lasso Coefficients (Wine Quality)")
plt.ylabel("Coefficient value")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# -----------------------------
# 5) Residual plot
# -----------------------------
plt.figure(figsize=(6, 4))
plt.scatter(y_pred, residuals, alpha=0.7)
plt.axhline(0, linewidth=1)
plt.title("Residual Plot (Residuals vs Predicted)")
plt.xlabel("Predicted quality")
plt.ylabel("Residual (Actual - Predicted)")
plt.tight_layout()
plt.show()
```

```python
# =========================
# COMMON CODE (Q58–Q61)
# One-time imports + reusable steps
# =========================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Lasso, Ridge
from sklearn.metrics import mean_squared_error, r2_score


def run_regression(file_path, target, model_type="lasso", alpha=0.1,
                   rename_map=None, drop_cols=None, sep=","):
    """
    Trains on COMPLETE dataset, plots coefficients + residual plot.
    model_type: "lasso" or "ridge"
    alpha: regularization strength
    rename_map: dict to rename columns
    drop_cols: list of columns to drop (e.g., IDs)
    """

    # 1) Load
    df = pd.read_csv(file_path, sep=sep)

    # 2) Rename if needed
    if rename_map:
        df.rename(columns=rename_map, inplace=True)

    # 3) Split X/y
    if target not in df.columns:
        raise ValueError(f"Target '{target}' not found. Available: {list(df.columns)}")

    X = df.drop(columns=[target])
    y = df[target]

    # 4) Drop unwanted columns (IDs etc.)
    if drop_cols:
        for c in drop_cols:
            if c in X.columns:
                X = X.drop(columns=[c])

    # 5) Preprocess: scale numeric + one-hot categorical
    num_cols = X.select_dtypes(include=["number"]).columns
    cat_cols = X.select_dtypes(exclude=["number"]).columns

    preprocess = ColumnTransformer(
        transformers=[
            ("num", StandardScaler(), num_cols),
            ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
        ],
        remainder="drop"
    )

    # 6) Choose model
    if model_type.lower() == "lasso":
        reg = Lasso(alpha=alpha, max_iter=10000)
        title_model = f"Lasso (alpha={alpha})"
    elif model_type.lower() == "ridge":
        reg = Ridge(alpha=alpha)
        title_model = f"Ridge (alpha={alpha})"
    else:
        raise ValueError("model_type must be 'lasso' or 'ridge'")

    model = Pipeline([
        ("prep", preprocess),
        ("reg", reg)
    ])

    # 7) Train on complete dataset
    model.fit(X, y)

    # 8) Predict + residuals
    y_pred = model.predict(X)
    residuals = y - y_pred

    print("\n==============================")
    print(f"File: {file_path}")
    print(f"Target: {target}")
    print(f"Model: {title_model}")
    print(f"RMSE: {np.sqrt(mean_squared_error(y, y_pred)):.4f}")
    print(f"R^2 : {r2_score(y, y_pred):.4f}")

    # 9) Coefficients (after preprocessing)
    feature_names = model.named_steps["prep"].get_feature_names_out()
    coefs = model.named_steps["reg"].coef_

    # 10) Plot coefficients
    plt.figure(figsize=(10, 4))
    plt.bar(feature_names, coefs)
    plt.axhline(0, linewidth=1)
    plt.title(f"Coefficients - {title_model}")
    plt.ylabel("Coefficient value")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

    # 11) Residual plot
    plt.figure(figsize=(6, 4))
    plt.scatter(y_pred, residuals, alpha=0.7)
    plt.axhline(0, linewidth=1)
    plt.title("Residual Plot (Actual - Predicted)")
    plt.xlabel("Predicted")
    plt.ylabel("Residual")
    plt.tight_layout()
    plt.show()

    return model


# =========================
# Q58) Housing (Lasso)
# Your file: Housing.csv
# Map: bedrooms -> num_rooms, stories -> age (because age not present)
# =========================
run_regression(
    file_path="Housing.csv",
    target="price",
    model_type="lasso",
    alpha=0.1,
    rename_map={"bedrooms": "num_rooms", "stories": "age"},
    drop_cols=None
)

# =========================
# Q59) Diabetes (Ridge)
# Your file: diabetes.csv
# Target is Outcome (because diabetes_progression not present)
# =========================
run_regression(
    file_path="diabetes.csv",
    target="Outcome",
    model_type="ridge",
    alpha=1.0,
    drop_cols=None
)

# =========================
# Q60) WineQuality (Lasso)
# If your CSV uses semicolon separator, set sep=";"
# =========================
run_regression(
    file_path="WineQuality.csv",
    target="quality",
    model_type="lasso",
    alpha=0.05,
    sep=","   # change to ";" if needed
)

# =========================
# Q61) HealthData (Lasso)
# Your file columns don't include health_score.
# Pick a numeric target (example: Cholesterol_Level)
# Drop Patient_ID (ID column)
# =========================
run_regression(
    file_path="HealthData.csv.csv",
    target="Cholesterol_Level",   # change if your teacher says another target
    model_type="lasso",
    alpha=0.05,
    drop_cols=["Patient_ID"]
)
```
