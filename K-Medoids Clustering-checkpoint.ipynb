<h3>9.Generate a random dataset of urban areas with features such as population density, average income, and
amenities, (min. 100 data). Apply K-medoids clustering to identify neighbourhoods with similar
characteristics. Visualize the clusters on a map</h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn_extra.cluster import KMedoids

# 1. Generate random data (100+)
n = 120
data = pd.DataFrame({
    "population_density": np.random.randint(5000,20000,n),
    "average_income": np.random.randint(20000,80000,n),
    "amenities": np.random.randint(5,50,n),
    "lat": np.random.uniform(10,30,n),
    "lon": np.random.uniform(70,90,n)
})

# 2. Apply K-Medoids
X = data[["population_density","average_income","amenities"]]
model = KMedoids(n_clusters=4, random_state=0)
data["cluster"] = model.fit_predict(X)

# 3. Map visualization
plt.scatter(data["lon"], data["lat"], c=data["cluster"])
plt.title("Urban Area Clusters")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()

# 4. Summary
print(data.groupby("cluster").mean())
```

<h3>10. Generate a random dataset containing customer information (e.g., age, spending score)(100 rows), apply
K-medoids clustering to segment customers into distinct groups. Visualize the clusters and summarize the
characteristics of each segment</h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn_extra.cluster import KMedoids

# 1. Generate customer data
n = 100
data = pd.DataFrame({
    "age": np.random.randint(18,65,n),
    "spending_score": np.random.randint(1,100,n)
})

# 2. Apply K-Medoids
X = data[["age","spending_score"]]
model = KMedoids(n_clusters=3, random_state=0)
data["cluster"] = model.fit_predict(X)

# 3. Plot clusters
plt.scatter(data["age"], data["spending_score"], c=data["cluster"])
plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.title("Customer Segments")
plt.show()

# 4. Segment summary
print(data.groupby("cluster").mean())
```
