<h3>Common Imports</h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

<h3>11.You are working with an e-commerce company that wants to segment its customers based on their
purchasing behaviour. Generate random dataset that includes features like total spending, frequency of
purchases, and average basket size. Develop a Python program to perform K-means clustering. Visualize
the clusters and summarize the characteristics of each segment.</h3>

```python
# 1) Generate a random customer dataset (realistic segments)
np.random.seed(42)
n1, n2, n3 = 200, 180, 140  # 3 underlying segments
# Segment A: high spenders, frequent, big baskets
seg1 = np.column_stack([
    np.random.normal(12000, 2500, n1),  # total_spending
    np.random.normal(35, 7, n1),        # purchase_frequency
    np.random.normal(1200, 250, n1)     # avg_basket_size
])

# Segment B: medium spenders, moderate frequency, medium baskets
seg2 = np.column_stack([
    np.random.normal(6000, 1500, n2),
    np.random.normal(18, 5, n2),
    np.random.normal(700, 150, n2)
])

# Segment C: low spenders, low frequency, small baskets
seg3 = np.column_stack([
    np.random.normal(2500, 800, n3),
    np.random.normal(7, 3, n3),
    np.random.normal(350, 120, n3)
])

X_raw = np.vstack([seg1, seg2, seg3])

# Keep values non-negative and sensible
X_raw[:, 0] = np.clip(X_raw[:, 0], 0, None)   # spending
X_raw[:, 1] = np.clip(X_raw[:, 1], 1, None)   # frequency >= 1
X_raw[:, 2] = np.clip(X_raw[:, 2], 50, None)  # basket size >= 50

df = pd.DataFrame(X_raw, columns=["total_spending", "purchase_frequency", "avg_basket_size"])
```

```python
# -----------------------------
# 2) Scale features (important for K-means)
# -----------------------------
scaler = StandardScaler()
X = scaler.fit_transform(df)
```

```python
# -----------------------------
# 3) K-means clustering
# -----------------------------
k = 3  # you can change this (e.g., use elbow method below)
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)
df["cluster"] = labels
```

```python
# -----------------------------
# 4) Visualize clusters (2D scatter plots)
# -----------------------------
plt.figure(figsize=(7, 5))
for c in sorted(df["cluster"].unique()):
    subset = df[df["cluster"] == c]
    plt.scatter(subset["total_spending"], subset["purchase_frequency"], label=f"Cluster {c}", alpha=0.7)

plt.title("Customer Segments (Total Spending vs Purchase Frequency)")
plt.xlabel("Total Spending")
plt.ylabel("Purchase Frequency")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(7, 5))
for c in sorted(df["cluster"].unique()):
    subset = df[df["cluster"] == c]
    plt.scatter(subset["purchase_frequency"], subset["avg_basket_size"], label=f"Cluster {c}", alpha=0.7)

plt.title("Customer Segments (Frequency vs Avg Basket Size)")
plt.xlabel("Purchase Frequency")
plt.ylabel("Average Basket Size")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# -----------------------------
# 5) Summarize characteristics of each segment
# -----------------------------
summary = df.groupby("cluster")[["total_spending", "purchase_frequency", "avg_basket_size"]].agg(
    count=("total_spending", "size"),
    mean_spending=("total_spending", "mean"),
    mean_frequency=("purchase_frequency", "mean"),
    mean_basket=("avg_basket_size", "mean"),
    median_spending=("total_spending", "median"),
    median_frequency=("purchase_frequency", "median"),
    median_basket=("avg_basket_size", "median"),
).round(2)

print("\n=== Segment Summary (per cluster) ===")
print(summary)

# Optional: label clusters from lowhigh spending for easier interpretation
order = summary["mean_spending"].sort_values().index.tolist()
label_map = {old: new for new, old in enumerate(order)}  # 0=lowest spending, 2=highest spending
df["cluster_ranked"] = df["cluster"].map(label_map)

print("\nCluster ranking by mean spending (0=lowest, higher=more valuable):")
print(df.groupby("cluster_ranked")[["total_spending", "purchase_frequency", "avg_basket_size"]].mean().round(2))
```

```python
# -----------------------------
# (Optional) Elbow method to choose K
# -----------------------------
inertias = []
Ks = range(2, 9)
for kk in Ks:
    km = KMeans(n_clusters=kk, random_state=42, n_init=10)
    km.fit(X)
    inertias.append(km.inertia_)
plt.figure(figsize=(6,4))
plt.plot(list(Ks), inertias, marker="o")
plt.title("Elbow Method")
plt.xlabel("K")
plt.ylabel("Inertia")
plt.tight_layout()
plt.show()
```

<h3>12) make_blobs (300 samples) + K from 2 to 7</h3>
<h4>Use make_blobs to generate 300 samples. Analyse the data and apply KMeans clustering and generate the
graph for better representation. Show what happens if you change n_clusters count from 2 to
7(2,3,4,5,6,7)?</h4>

```python
# Generate data
X, y = make_blobs(n_samples=300, centers=4, cluster_std=1.5, random_state=42)
```

```python
# Try cluster counts from 2 to 7
plt.figure(figsize=(15,8))

for k in range(2,8):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)

    plt.subplot(2,3,k-1)
    plt.scatter(X[:,0], X[:,1], c=labels)
    plt.title(f"K = {k}")

plt.tight_layout()
plt.show()
```

<h3>13.Grocery store transaction clustering</h3>
<h4>A grocery store wants to analyze customer purchase patterns to improve product placements. Generate
random transaction data (products purchased together), apply K-means clustering to group similar
transactions.</h4>

```python
# Random transaction data
np.random.seed(42)

data = pd.DataFrame({
    "milk": np.random.randint(0,2,100),
    "bread": np.random.randint(0,2,100),
    "butter": np.random.randint(0,2,100),
    "fruits": np.random.randint(0,2,100),
    "snacks": np.random.randint(0,2,100)
})
```

```python
# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
data["cluster"] = kmeans.fit_predict(data)

# Plot using 2 features for visualization
plt.scatter(data["milk"], data["bread"], c=data["cluster"])
plt.xlabel("Milk")
plt.ylabel("Bread")
plt.title("Customer Transaction Clusters")
plt.show()
```

```python
# Cluster summary
print(data.groupby("cluster").mean())
```

Interpretation

Clusters may represent:

Cluster 0  Milk + Bread buyers

Cluster 1  Snacks lovers

Cluster 2  Fruits + Healthy buyers

Store can place related products together.

<h3>14) 35 random points + different cluster colors</h3>
<h4>Generate random dataset with 35 points. Apply KMeans clustering and generate the plot with each cluster
points with different colours. What is the impact of varying cluster count? Justify your understanding with
suitable graphs</h4>

```python
# Generate 35 points
X = np.random.rand(35,2)
```

```python
# Try different cluster counts
plt.figure(figsize=(12,6))

for k in [2,3,4,5]:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)

    plt.subplot(2,2,k-1)
    plt.scatter(X[:,0], X[:,1], c=labels)
    plt.title(f"K = {k}")

plt.tight_layout()
plt.show()
```


Explanation for exam:

Increasing clusters divides data into smaller groups.

Too few clusters  Underfitting

Too many clusters  Overfitting

Best K chosen using Elbow method.

<h1> Use Elbow Method</h1>

```python
inertia = []
for k in range(1,10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.plot(range(1,10), inertia, marker='o')
plt.xlabel("K")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```
